# 計算IE遞延五年的RD費用

```python
RD_5 = pd.read_csv('IE_year_V2.csv')
RD5_cp_list = list(set(RD_5['firm']))
```

```python
col_name = ['firm','year','RD_expense']
c_df = RD_5[col_name]
```

```python
c_df = c_df.pivot(index='firm',columns='year',values='RD_expense')
c_df_V2 =  c_df.reset_index() 
c_df_V2 = pd.DataFrame(c_df_V2)
```

```python
c_df_V2
```

```python
df1= pd.DataFrame()
df2= pd.DataFrame()

for c in RD5_cp_list: 
    df = c_df_V2[c_df_V2["firm"]==c]

    weights = [0.2, 0.4, 0.6, 0.8, 1]
    attributes_2016 = [2011,2012, 2013, 2014,2015]
    attributes_2017 = [2012, 2013, 2014, 2015, 2016]
    attributes_2018 = [2013, 2014, 2015, 2016, 2017]
    attributes_2019 = [2014, 2015, 2016, 2017, 2018]
    attributes_2020 = [2015, 2016, 2017, 2018, 2019]
    attributes_2021 = [2016, 2017, 2018, 2019, 2020]

    df['RD_5_2016'] = df[attributes_2016].mul(weights).sum(1)
    df['RD_5_2017'] = df[attributes_2017].mul(weights).sum(1)
    df['RD_5_2018'] = df[attributes_2018].mul(weights).sum(1)
    df['RD_5_2019'] = df[attributes_2019].mul(weights).sum(1)
    df['RD_5_2020'] = df[attributes_2020].mul(weights).sum(1)
    df['RD_5_2021'] = df[attributes_2021].mul(weights).sum(1)

    df2 = df2.append(df,ignore_index=True)
```

```python
df2
```

```python
df2.to_csv('RD_5_V3.csv',index=False)
```

# 計算IO 前五年原創分數相加並取mean

```python
Ori_5 = pd.read_csv('~/Desktop/論文/資料集/dataset/tidy_up/專利資料/IO_Score_sum_V1.csv')
Ori_5_cp_list = list(set(Ori_5['firm']))
```

```python
Ori_5
```

```python
Ori_5_df = Ori_5.pivot(index='firm',columns='year',values='IO_Score_sum')
Ori_5_df_V2 =  Ori_5_df.reset_index() 
Ori_5_df_V2 = pd.DataFrame(Ori_5_df_V2)
Ori_5_df_V2 =Ori_5_df_V2.fillna(value=0)
```

```python
Ori_5_df_V2
```

```python
df1= pd.DataFrame()
df2= pd.DataFrame()

for c in Ori_5_cp_list: 
    df = Ori_5_df_V2[Ori_5_df_V2["firm"]==c]

    # weights = [0.2, 0.4, 0.6, 0.8, 1]
    attributes_2015 = [2010,2011, 2012, 2013,2014]
    attributes_2016 = [2011,2012, 2013, 2014,2015]
    attributes_2017 = [2012, 2013, 2014, 2015, 2016]
    attributes_2018 = [2013, 2014, 2015, 2016, 2017]
    attributes_2019 = [2014, 2015, 2016, 2017, 2018]
    attributes_2020 = [2015, 2016, 2017, 2018, 2019]
    attributes_2021 = [2016, 2017, 2018, 2019, 2020]


    df['Ori_5_2015'] = df[attributes_2015].mean(1)
    df['Ori_5_2016'] = df[attributes_2016].mean(1)
    df['Ori_5_2017'] = df[attributes_2017].mean(1)
    df['Ori_5_2018'] = df[attributes_2018].mean(1)
    df['Ori_5_2019'] = df[attributes_2019].mean(1)
    df['Ori_5_2020'] = df[attributes_2020].mean(1)
    df['Ori_5_2021'] = df[attributes_2021].mean(1)

    df2 = df2.append(df,ignore_index=True)
```

```python
df2
```

```python
df2.to_csv('Ori_5_V1.csv',index=False)
```
