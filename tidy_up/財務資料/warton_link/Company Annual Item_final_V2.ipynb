# Descibe2:每間公司每年的變化量 【財務資料_每年變動.csv】

```python
import pandas as pd
import numpy as np
```

```python
describe_date = pd.read_csv("財務資料表.csv", index_col=["Company Name","datadate"]) 
describe_date = describe_date.reindex(columns =["Ticker", "gvkey", "sic","\tResearch&Development_Expense","Assets_Total", "Net_Income(Loss)", "Earnings Before Interest and Taxes", "Operating_Net_Cash_Flow", "Earnings_Per_Share_from_Operations", "Stockholders_Equity_Total" ])  
describe_date
```

```python
data_new = describe_date.reset_index(drop=False) #將固定的index拿掉
data_new
# data_new[0:25]
```

```python
company_ls = list(set(data_new["Company Name"])) #全部公司名單
c_ls=[]
for c in company_ls: 
    df = data_new[data_new["Company Name"]==c]
    df.reset_index(drop=True,inplace=True)

    for i in range(1,len(df)):
        num_ls = list(df.iloc[i,5:]-df.iloc[i-1,5:]/df.iloc[i-1,5:]) #後一項減前一項
        num_ls.insert(0,df["Company Name"][0]) #加入公司名稱
        num_ls.insert(1,df.datadate[i][0:4]+"-"+df.datadate[i-1][0:4]) #加入日期區間
        num_ls.insert(2,df["Ticker"][0]) #加入Ticker
        num_ls.insert(3,df["gvkey"][0]) #加入gvkey
        num_ls.insert(4,df["sic"][0]) #加入sic       
        c_ls.append(num_ls)
        
```

```python
col_ls = list(data_new.columns)
df_test = pd.DataFrame(c_ls, columns = col_ls, dtype = float)
df_test = df_test.sort_values(by=['gvkey'])
df_test
```

```python
df_test.to_csv("財務資料_每年變動.csv",index=False)
```
