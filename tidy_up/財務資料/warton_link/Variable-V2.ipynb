## 計算RD系列

```python
variables['RD/ME'] = variables['Research&Development_Expense']/variables['Stockholders_Equity_Total'] 
variables['RDI'] = variables['Research&Development_Expense']/variables['sales'] #RD/Sales
variables['RDD'] = variables['Research&Development_Expense']/variables['Operating_Profit'] #RD/operating profit
variables
```

```python
# variables.to_csv("財務資料_variables.csv",index=False)
```

## 計算 CapEx, R&D, Advertising, SG&A, OCF, Leverage

```python
print(variables.columns)
```

```python
variables['MS_CapEx'] = variables['capital_expenditure']/variables['Assets_Total']
variables['MS_R&D'] = variables['Research&Development_Expense']/variables['Assets_Total']
variables['MS_Advertising'] = variables['Ad_expense']/variables['Assets_Total']
variables['MS_SG&A'] = variables['SG&A']/variables['Assets_Total']
variables['MS_OCF'] = variables['Operating_Net_Cash_Flow']/variables['Assets_Total']
variables['MS_Leverage'] = variables['total_debt']/variables['Assets_Total']
variables
```

## 計算ROA/ROE & OCF normalized

```python
import numpy as np
```

```python
variables['ROA'] = variables['Net_Income(Loss)']/variables['Assets_Total']
variables['ROE'] = variables['Net_Income(Loss)']/variables['Stockholders_Equity_Total']
variables
```

```python
variables['OCF_norm'] = variables['Operating_Net_Cash_Flow']/np.linalg.norm(variables['Operating_Net_Cash_Flow'])
```

## 計算資產成長率

```python
company_ls = list(set(variables["Company Name"])) #全部公司名單
```

```python
df1= pd.DataFrame()
df2= pd.DataFrame()
# i=0
for c in company_ls: 
    df = variables[variables["Company Name"]==c]
#     print(i)
#     i=i+1
    df['Assets_G']=(df['Assets_Total']-df['Assets_Total'].shift(1))/df['Assets_Total']
    df['delta_ROA']=(df['ROA']-df['ROA'].shift(1))
    df['delta_OCF']=(df['Operating_Net_Cash_Flow']-df['Operating_Net_Cash_Flow'].shift(1))
    df['ROA_2021']=df['ROA'].shift(-1)
    df['OCF_2021']=df['Operating_Net_Cash_Flow'].shift(-1)
    df['OCF_norm_2021']=df['OCF_norm'].shift(-1)
    df['ROE_2021']=df['ROE'].shift(-1)

    
    
    df1=pd.DataFrame(df)
#     print(df1)
    df2 = df2.append(df1,ignore_index=True)
```

```python
df2
```

##  Drop Asset_G==0 [去除2010年]---時間:2011~2020(10年)

```python
clean_data = df2[(df2.Assets_G.isna() == False )& (df2.delta_ROA.isna() == False) & 
                 (df2.delta_OCF.isna() == False ) & (df2.ROA_2021.isna() == False ) &
                (df2.OCF_2021.isna() == False )&(df2.OCF_norm_2021.isna() == False ) &
                (df2.ROE_2021.isna() == False )]
```
